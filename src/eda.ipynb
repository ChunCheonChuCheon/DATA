```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../asset/c4_general_cafeteria_chuncheon_opened_SS-HJ_addLatiLongti_menu-price_addedLabel_v3.csv', encoding='utf-8-sig')
df
```

```python
plt.rc('font', family='AppleGothic')
sns.countplot(x='label', data=df, palette='rainbow')
plt.title('DB에 존재하는 식당의 라벨분포(n=852)')
plt.xlabel('라벨분포')
plt.ylabel('식당 수')
plt.xticks(rotation=45)
plt.show()

# 효자동과 석사동 class에 따라 식당의 수를 나타냄 (레전드 : 효자동, 석사동)
plt.rc('font', family='AppleGothic')
sns.countplot(x='label', data=df, hue='동/면', palette='rainbow')
plt.title('효자동과 석사동 식당의 라벨분포(n=852)')
plt.xlabel('라벨분포')
plt.ylabel('식당 수')
plt.xticks(rotation=45)
plt.show()
```
