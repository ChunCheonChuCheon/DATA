```python
import pandas as pd

df = pd.read_csv('../asset/c4_general_cafeteria_chuncheon_opened_SS-HJ_addLatiLongti_menu-price_addedLabel.csv', encoding='utf-8')
```

```python
# index_cafeteria를 기준으로 그룹화하고 리스트로 변환
grouped = df.groupby('index_cafeteria').agg(list)

# NaN 값이 포함된 리스트를 정리
cleaned_grouped = grouped.applymap(lambda x: [i for i in x if pd.notna(i)])

cleaned_grouped
```

```python
# cleaned_grouped['cafeteria_name']의 0번째 인덱스에 해당하는 값 추출(원본에 적용)
df['cafeteria_name'] = cleaned_grouped['cafeteria_name'].apply(lambda x: x[0])
```

```python
cleaned_grouped
```

```python
cleaned_grouped['cafeteria_name'] = cleaned_grouped['cafeteria_name'].apply(lambda x: x[0])
cleaned_grouped
```

```python
df_org = pd.read_csv('../asset/c4_general_cafeteria_chuncheon_opened_SS-HJ_addLatiLongti.csv', encoding='utf-8')
```

```python
# '사업장명'열의 이름을 'cafeteria_name'으로 변경
df_org.rename(columns={'사업장명': 'cafeteria_name'}, inplace=True)
```

```python
# df_org과 cleaned_grouped를 합침(이떄 합치는 기준은 index_cafeteria)

df_merged = pd.merge(df_org, cleaned_grouped, on='index_cafeteria')
df_merged

```

```python
# 모든행을 기준으로 cafeteria_name_x와 cafeteria_name_y가 같은지 비교
df_merged['cafeteria_name_x'] == df_merged['cafeteria_name_y']
```

```python
# df_merged['cafeteria_name_x']의 이름을 cafeteria_name로 바꾸고
# df_merged['cafeteria_name_y']열을 삭제
df_merged.rename(columns={'cafeteria_name_x': 'cafeteria_name'}, inplace=True)
df_merged.drop('cafeteria_name_y', axis=1, inplace=True)
```

```python
#csv로 저장
df_merged.to_csv('../asset/c4_general_cafeteria_chuncheon_opened_SS-HJ_addLatiLongti_menu-price_addedLabel_v2.csv', encoding='utf-8', index=False)
```
